```python
import xml.etree.ElementTree as ET
import json
```

```python
import requests
from requests.auth import HTTPBasicAuth
```

```python
def call_sics(file_url):
    season_to_search = 2022
    competition_to_search = 'SERIE A'
    competition_dict = {'SERIE A':'2'}

    username_ = 'riccardogobbo17@gmail.com'
    pwd_ = 'YTCWC)FyUfuEfv7'

    rankings_link = file_url + '?seasonId=' + str(season_to_search) + '&competitionId=' + str(competition_dict[competition_to_search]) + '&lang=it'

    headers = {'Accept': 'application/xml'}

    # Effettua la richiesta HTTP
    response = requests.get(rankings_link, headers=headers, auth=HTTPBasicAuth(username_, pwd_))

    # Verifica lo stato della richiesta
    if response.status_code == 200:
        xml_data = response.text
    else:
        print("Errore durante il download dei dati:", response.status_code)
        xml_data = None
    return xml_data
```

```python
xmldata = call_sics('http://server.sics.it/livefeeds/x05.xml')
teamdata = call_sics('http://server.sics.it/livefeeds/s01.xml')
```

```python
root = ET.fromstring(teamdata)
```

```python
teams = {}
for item in root.findall("teams"):
    name = item.find('name').text
    logo = item.find('logo').text
    teams[name.upper()] = {'name': name.lower().title(), 'logo': logo}
```

```python
root = ET.fromstring(xmldata)
```

```python
tree = ET.parse('x05 ranking and results.xml')
root = tree.getroot()
```

```python
ranking = []
for item in root.find("fixtureStat").find("ranking").findall("team"):
    name = item.get('name')
    points = int(item.find('totpoints').text)
    ranking.append({'team': teams[name.upper()], 'points': points})
```
